```python
import requests
import pandas as pd
import time
import bs4
from bs4 import BeautifulSoup as bs
from selenium import webdriver
from datetime import date
```

```python
#dataframe to store results
#job title, company, location, salary, job desc & page of scrape
df = pd.DataFrame(columns = ['Job Title', 'Company', 'Location', 'Salary', 'Job Desc', 'Page no.'])
```

```python
#Use Chrome webdriver
driver = webdriver.Chrome()

#for pages 1-10, in increments of 10
for i in range(0,1010,10):
    
    #get the page
    driver.get('https://sg.indeed.com/jobs?q=quantitative&start='+str(i))
    driver.implicitly_wait(4)
    
    #get all divs with result in it
    jobs = driver.find_elements_by_class_name('result')
    
    for job in jobs:
        result_html = job.get_attribute('innerHTML')
        soup = bs(result_html, 'html.parser')

        #job title
        try:
            title = soup.find('a', class_='jobtitle').text.strip()
        except:
            title = 'None'

        #company
        try:
            comp = soup.find(class_='company').text.strip()
        except Exception as e:
            comp = None

        #location
        try:
            location = soup.find(class_='location').text.strip()
        except Exception as e:
            location = None

        #job salary
        try:
            salary = soup.find(class_='salaryText').text.strip()
        except Exception as e:
            salary = None

        #job desc
        sum_div = job.find_elements_by_class_name('summary')[0]
        
        try:
            sum_div.click()
        except:
            close_button = driver.find_elements_by_class_name('popover-x-button-close')[0]
            close_button.click()
            sum_div.click()
        job_desc = driver.find_element_by_id('vjs-desc').text.strip()
        
        driver.implicitly_wait(4)
        

        df = df.append({'Job Title': title, 'Company':comp, 'Location': location, 'Salary':salary, 
                        'Job Desc':job_desc, 'Page no.':int(i/10)}, ignore_index=True)
```
